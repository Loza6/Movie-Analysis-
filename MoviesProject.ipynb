Step 4: Explore Data with Visualizations

Create visualizations to explore the dataset further. You can use libraries like Matplotlib or Seaborn to generate visualizations.
Create histograms for numeric variables to understand their distributions.
Create scatter plots to visualize relationships between variables.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV file into a DataFrame
df = pd.read_csv('movies.csv')

#relevant columns for visualization
columns_to_visualize = ['YEAR', 'RATING', 'VOTES', 'RunTime', 'Gross']

# Dropping rows with missing values in selected columns
df_cleaned = df.dropna(subset=columns_to_visualize)

# histograms for numeric columns
plt.figure(figsize=(14, 10))
for i, column in enumerate(columns_to_visualize, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df_cleaned[column], kde=True, color='skyblue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
plt.tight_layout()
plt.show()

# scatter plots to visualize relationships
sns.set(style='whitegrid')
sns.pairplot(df_cleaned[columns_to_visualize])
plt.suptitle('Pair Plot of Numeric Columns', y=1.02)
plt.show()
```

Step 6: Exploratory Data Analysis (EDA)

Perform in-depth data exploration by creating more advanced visualizations, such as box plots, pair plots, and correlation matrices.
Identify trends and correlations between variables. Document your findings and insights.
Use EDA techniques to answer

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV file into a DataFrame
df = pd.read_csv('movies.csv')

# relevant columns for EDA
columns_to_eda = ['YEAR', 'RATING', 'VOTES', 'RunTime', 'Gross']

# Dropping rows with missing values in selected columns
df_cleaned = df.dropna(subset=columns_to_eda)

#box plots for numeric columns
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_cleaned, palette='Set2')
plt.title('Box Plots of Numeric Columns')
plt.xticks(rotation=45)
plt.show()

# Creating a pair plot for numeric columns
sns.pairplot(df_cleaned[columns_to_eda])
plt.suptitle('Pair Plot of Numeric Columns', y=1.02)
plt.show()

# correlation matrix
correlation_matrix = df_cleaned[columns_to_eda].corr()

#heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
